Compare currently analysing candidate with a peer candidate

```python
from github import Github

# Authenticate with GitHub
access_token = "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"
g = Github(access_token)

# Fetch user information
current_username = "Maldeniya99"
current_user = g.get_user(current_username)

# Fetch another user information
other_username = "IT20207854"
other_user = g.get_user(other_username)

# Calculate language proficiency for the current user
current_language_proficiency = {}
for repo in current_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in current_language_proficiency:
            current_language_proficiency[language] += value
        else:
            current_language_proficiency[language] = value

# Calculate language proficiency for the other user
other_language_proficiency = {}
for repo in other_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in other_language_proficiency:
            other_language_proficiency[language] += value
        else:
            other_language_proficiency[language] = value

# Compare language proficiency
common_languages = set(current_language_proficiency.keys()) & set(other_language_proficiency.keys())

print("Language Proficiency Comparison:")
for language in common_languages:
    current_percentage = (current_language_proficiency[language] / sum(current_language_proficiency.values())) * 100
    other_percentage = (other_language_proficiency[language] / sum(other_language_proficiency.values())) * 100
    print(f"- {language}:")
    print(f"    Current User: {current_percentage:.2f}%")
    print(f"    Other User: {other_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Calculate language proficiency for the current user
current_language_proficiency = {}
for repo in current_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in current_language_proficiency:
            current_language_proficiency[language] += value
        else:
            current_language_proficiency[language] = value

# Calculate language proficiency for the other user
other_language_proficiency = {}
for repo in other_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in other_language_proficiency:
            other_language_proficiency[language] += value
        else:
            other_language_proficiency[language] = value

# Compare language proficiency
common_languages = list(set(current_language_proficiency.keys()) & set(other_language_proficiency.keys()))

# Prepare data for visualization
current_percentages = [(current_language_proficiency[lang] / sum(current_language_proficiency.values())) * 100 for lang in common_languages]
other_percentages = [(other_language_proficiency[lang] / sum(other_language_proficiency.values())) * 100 for lang in common_languages]

# Plot the language proficiencies
plt.figure(figsize=(10, 6))
plt.bar(common_languages, current_percentages, label='Current User')
plt.bar(common_languages, other_percentages, label='Other User', alpha=0.5)

# Add labels and title
plt.xlabel('Language')
plt.ylabel('Percentage')
plt.title('Language Proficiency Comparison')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Current candidate VS Peer candidate Comparison ( Knowldge Graph )

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
graph = nx.DiGraph()

# Add nodes for users
graph.add_node("Current User")
graph.add_node("Other User")

# Calculate language proficiency for the current user
current_language_proficiency = {}
for repo in current_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in current_language_proficiency:
            current_language_proficiency[language] += value
        else:
            current_language_proficiency[language] = value

# Calculate language proficiency for the other user
other_language_proficiency = {}
for repo in other_user.get_repos():
    languages = repo.get_languages()
    for language, value in languages.items():
        if language in other_language_proficiency:
            other_language_proficiency[language] += value
        else:
            other_language_proficiency[language] = value

# Compare language proficiency
common_languages = set(current_language_proficiency.keys()) & set(other_language_proficiency.keys())

# Add nodes for common languages
for language in common_languages:
    graph.add_node(language)

# Add edges for language proficiency
for language in common_languages:
    current_percentage = (current_language_proficiency[language] / sum(current_language_proficiency.values())) * 100
    other_percentage = (other_language_proficiency[language] / sum(other_language_proficiency.values())) * 100
    graph.add_edge("Current User", language, weight=current_percentage)
    graph.add_edge("Other User", language, weight=other_percentage)

# Define layout for the graph
pos = nx.spring_layout(graph)

# Extract weights for edge labels
edge_labels = nx.get_edge_attributes(graph, "weight")

# Plot the graph
plt.figure(figsize=(10, 6))
nx.draw_networkx(graph, pos, with_labels=True, node_color="lightblue", node_size=1000, font_size=10, alpha=0.8)
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)

# Add title
plt.title("Language Proficiency Comparison")

# Show the plot
plt.show()
```
